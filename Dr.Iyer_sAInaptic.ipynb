```python
import spacy
from spacy.matcher import PhraseMatcher

nlp = spacy.load("en_core_web_sm")
matcher = PhraseMatcher(nlp.vocab)
terms = ["respiration", "absence of oxygen", "without oxygen"]

# Only run nlp.make_doc to speed things up
patterns = [nlp.make_doc(text) for text in terms]
matcher.add("TerminologyList", patterns)

doc = nlp("breathing without oxygen "
          "respiration when there is no oxygen present in the air"
          "respiration in the absence of oxygen"
          "when respiration takes place with lack of oxygen, it is known as anaerobic respiration")
matches = matcher(doc)
for match_id, start, end in matches:
    span = doc[start:end]
    print(span.text)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

from spacy.matcher import Matcher

m_tool = Matcher(nlp.vocab)
```

```python
# Creatung different rules to identify answers written by students

p1 = [{'LOWER': 'nooxygen'}]
p2 = [{'LOWER': 'respiration'}, {'IS_PUNCT': True}, {'LOWER': 'when'},  {'IS_PUNCT': True}, {'LOWER': 'oxygen'}]
p3 = [{'LOWER': 'breathing'}, {'LOWER': 'without'}, {'LOWER': 'oxygen'}]
p4 =  [{'LOWER': 'breathing'}, {'LOWER': 'without-oxygen'}]
```

```python
m_tool.add('QBF', None, p1, p2, p3, p4)
```

```python
sentence = nlp(u'breathing without oxygen. respiration when there is no oxygen present in the air. \
               respiration in the absence of oxygen. when respiration takes place in the lack of oxygen')
```

```python
phrase_matches = m_tool(sentence)
print(phrase_matches )
```

```python
for match_id, start, end in phrase_matches:
    string_id = nlp.vocab.strings[match_id]  
    span = sentence[start:end]                   
    print(match_id, string_id, start, end, span.text)
```
