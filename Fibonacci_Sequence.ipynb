# Fibonacci Sequence in Python

---

In just 5 lines of code, you can implement the Fibonacci Sequence. Pretty Simple right?

```python
def fibonacci(n):
  a,b = 0,1

  for i in range(n):
    a,b = b, a + b
  return a
```

```python
print(fibonacci(0))
```

```python
print(fibonacci(1))
```

```python
print(fibonacci(2))
```

```python
print(fibonacci(3))
```

```python
print(fibonacci(4))
```
